```python
fig = go.Figure(data = go.Scatter(x=df['Distance'], y = df['DepTime'], mode='markers' , marker=dict(color='red')))

fig.update_layout(title='Scatter plot for Distance vs Departure time',xaxis_title = 'Distance', yaxis_title='Departure time')
fig.show()
```

```python
line_data = df.groupby('Month')['ArrDelay'].mean().reset_index()
```

```python
fig = go.Figure(data = go.Scatter(x=line_data['Month'], y = line_data['ArrDelay'], mode='lines' , marker=dict(color='green')))

fig.update_layout(title='Monthly average delay plot for an year', xaxis_title = 'Months', yaxis_title='Delay of flights')
fig.show()
```

```python
bar_data = df.groupby('DestState')['Flights'].sum().reset_index()
```

```python
fig = pr.bar(bar_data,x= 'DestState', y = 'Flights', title = 'Total number of flights from each state')

fig.show()
```

```python

bub_data = df.groupby('Reporting_Airline')['Flights'].sum().reset_index()
```

```python
fig_bub = pr.scatter(bub_data,x='Reporting_Airline',y='Flights',title='Total number of flights per reporting airlines',size="Flights",hover_name='Reporting_Airline',size_max=60)

fig_bub.show()
```
